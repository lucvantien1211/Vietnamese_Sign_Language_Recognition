```python
import cv2
import matplotlib.pyplot as plt
import sys
import os

sys.path.append(os.path.abspath(".."))
from src.data_utils import get_sample_frames
from src.plot_utils import plot_sample_frames
```

```python
example_video_path = "../data/dataset/train/Ăn/148050.mp4"

cap = cv2.VideoCapture(example_video_path)
```

```python
cap.isOpened()
```

```python
print("FPS:", cap.get(cv2.CAP_PROP_FPS))
print("Total frames:", int(cap.get(cv2.CAP_PROP_FRAME_COUNT)))
print("Width:", int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)))
print("Height:", int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
```

```python
cap.release()
```

```python
sample_frames = get_sample_frames(example_video_path)
```

```python
plt.figure(figsize=(12, 6))

for idx, frame in enumerate(sample_frames):
    plt.subplot(1, 5, idx+1)
    plt.imshow(frame)
    plt.axis("off")

plt.show()
```
